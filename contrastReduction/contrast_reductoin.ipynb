```python
import cv2
from matplotlib import pyplot as plt
import PIL as pillow
import numpy as np
```

```python
image = cv2.imread('luffy.png')
plt.imshow(image)
```

```python
imagem = cv2.bitwise_not(image) #using bitwise not over the image
plt.imshow(imagem)
cv2.imwrite('output.jpeg', imagem)
```
