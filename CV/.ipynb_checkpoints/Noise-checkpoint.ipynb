```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
noise = np.random.randint(low=0,high=2,size=(500,500))
```

```python
noise.shape
```

```python
noise
```

```python
plt.imshow(noise, cmap='gray')
```

```python
noise=noise*255
```

```python
noise
```

```python
img = np.zeros(shape=(500,500), dtype=np.int32)
plt.imshow(img, cmap='gray')
```

```python
cv2.circle(img, center=(250,250), radius=50, color=255, thickness=5)
plt.imshow(img,cmap='gray')
```

```python
img = img+noise
```

```python
plt.imshow(img,cmap='gray')
```

```python
noise1 = np.random.randint(low=0,high=2,size=(500,500))
```

```python
noise1
```

```python
noise1=noise1*-255
```

```python
noise1
```

```python
img1 = np.zeros(shape=(500,500), dtype=np.int32)
cv2.circle(img1,center=(250,250), radius=50,color=255,thickness=5)
plt.imshow(img1,cmap='gray')
```
