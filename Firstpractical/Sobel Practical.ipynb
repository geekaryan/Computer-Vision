```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('luffy.png', 0)  # Read the image in grayscale
```

```python
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=21)  # Sobel operator for x-direction
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=21)  # Sobel operator for y-direction
```

```python
edges = cv2.add(np.absolute(sobelx), np.absolute(sobely))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2) 
plt.imshow(edges, cmap='gray')
plt.title('Edges')

plt.show()
```
