```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('luffy.png', 0)  # Read the image in grayscale
```

```python
edges = cv2.Canny(image, 80, 80)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edges')

plt.show()
```
